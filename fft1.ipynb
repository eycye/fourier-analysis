```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

# Number of samplepoints
N = 256
# sample spacing
T = 4/N
omega = np.pi/(2.0*T)

l = np.linspace(-omega,omega, 5000)
u = 1/np.sqrt(2*np.pi)*(1/6)*1/(2j)*(1/(1+1j*l-6j)-1/(1+1j*l+6j))

x = np.linspace(0.0, N*T, N)
y = 1/6*np.sin(6*x)*(np.e**(-x))
yf = scipy.fftpack.fft(y)
xf = np.linspace(-omega,omega,N//2)

fig, ax = plt.subplots(figsize=(20,6))
ax.plot(l,np.abs(u),label='$\^{u}(\lambda)$')

for i in range(1, N//2+2):
    pt = 2*(i-1)*np.pi/(T*N)
    ax.plot(pt,T*np.abs(yf[i-1])/np.sqrt(2*np.pi),marker='.',color='r')
for i in range(N//2+2, N+1):
    pt = 2*(i-1-N)*np.pi/(T*N)
    ax.plot(pt,T*np.abs(yf[i-1])/np.sqrt(2*np.pi),marker='.',color='r')
plt.legend(fontsize=15)
plt.title("T=4, N=256",fontsize=30)
plt.show()
```
