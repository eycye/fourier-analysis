```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def partial1(start_x, end_X, N, points):
    x = np.linspace(start_x, end_X, points)
    y = []
    for xx in x:
        point = 0
        for n in range(1,N+1):
            point += np.power(-1, n) * np.cos(n * xx) * 4 / np.power(n, 2)
        y.append(point)

    for i in range(len(y)):
        y[i] += np.pi ** 2 /3
    return x, y
```

```python
q1_partial_sums = [1, 2, 5, 7]

fig, axs = plt.subplots(len(q1_partial_sums), sharex=True, sharey=True, figsize=(5,9))

lower, upper = -np.pi, np.pi

x = np.linspace(lower, upper, 500)
for i,N in enumerate(q1_partial_sums):
    axs[i].plot(x, x**2, label="f(x)=x^2")
    axs[i].plot(*partial1(lower, upper, N, 500), label="N={}".format(N))
    axs[i].grid()
    axs[i].legend(loc='best')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def partial2(start_x, end_X, N, points):
    x = np.linspace(start_x, end_X, points)
    y = []
    for xx in x:
        point = 0
        for n in range(1, N+1, 2):
            point += 2 * np.power(-1, (n-1)/2) * np.cos(np.pi * xx * n) / (n * np.pi)
        y.append(point)

    for i in range(len(y)):
        y[i] += 1/2
    return x, y
```

```python
q2_partial_sums = [5,10,20,40]

fig, axs = plt.subplots(len(q1_partial_sums), sharex=True, sharey=True, figsize=(5,9))

lower, upper = -1,1

x = np.linspace(lower, upper, 500)

def f_q2(x):
    return [1 if xx > -0.5 and xx <= 0.5 else 0 for xx in x]
    
for i,N in enumerate(q2_partial_sums):
    axs[i].plot(x, f_q2(x), label="f(x)")
    axs[i].plot(*partial2(lower, upper, N, 500), label="N={}".format(N))
    axs[i].grid()
    axs[i].legend(loc='best')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
